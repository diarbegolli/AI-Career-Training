```python
# Import the required libraries
import pandas as pd                # For data analysis
import seaborn as sns              # For datasets and visuals
import matplotlib.pyplot as plt    # For plotting graphs

# Load the Titanic dataset directly from seaborn's library
df = sns.load_dataset('titanic')

# Show the first 5 rows of the dataset
df.head()  # This gives a quick preview of what the data looks like


# Show how many rows and columns the dataset has
df.shape  # (891, 15)

# List all column names
df.columns

# Display basic information: column types, null values, etc.
df.info()

# Show statistical summary (mean, std, etc.) for numerical columns
df.describe()

# Count missing values in each column
df.isnull().sum()


df['age'].mean()  # Mean age of all passengers
df['survived'].value_counts()
# 0 = Died, 1 = Survived
df['survived'].value_counts(normalize=True) * 100
# Now shows percentages
df.groupby('sex')['survived'].mean()
# Average survival rate for males and females
df.sort_values('fare', ascending=False).head(5)
df['pclass'].value_counts()
# How many people were in 1st, 2nd, 3rd class
df.groupby('alone')['survived'].mean()
# Compares survival rates for people alone vs. not alone


df[df['age'] > 60]
df[(df['sex'] == 'female') & (df['survived'] == 1)]
df[(df['pclass'] == 3) & (df['age'] < 18)]
df[df['embarked'] == 'C']

#***VISUALIZATION PART***

%matplotlib inline
sns.set(style = "darkgrid")

#This is a bar plot showing the number of passengers for each value in the 'survived' column
sns.countplot(x='survived',data = df)
#Adds a title to your chart
plt.title('Survival Count')
#Changes the x-axis labels from [0, 1] to readable text ['Died', 'Survived']
plt.xticks([0,1],['Died','Survived'])
#Adds clear axis labels so your chart makes sense
plt.ylabel('Number of Passengers')
plt.xlabel('Survival Status')
#Tells matplotlib: Display the final plot now
plt.show()

sns.countplot(x='sex' , hue='survived' , data=df)
plt.title('Survival by Gendre')
plt.legend(title='Survived',labels=['No','Yes'])
plt.show()


sns.histplot(data=df, x='age', hue='survived', bins=20, kde=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

sns.histplot(data=df, x='fare', hue='survived', bins=40, kde=True)
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare ($)')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()






```
