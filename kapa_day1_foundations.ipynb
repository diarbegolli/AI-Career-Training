```python
# range 1 to 100 print based on mod
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("AI Bootcamp")
    elif i % 3 == 0:
        print("AI")
    elif i % 5 == 0:
        print("Bootcamp")
    else:
        print(i)

scores = [75, 88, 92, 67, 100, 81]

# find min,max, average from a list
stats = {
    "min": min(scores),
    "max": max(scores),
    "avg": sum(scores)/len(scores),
    "count_above_80": sum(1 for s in scores if s > 80)
}
print(stats)

def describe_number(n):
    # Even or odd
    if n % 2 == 0:
        parity = "Even"
    else:
        parity = "Odd"

    # Prime or composite
    if n < 2:
        primality = "Neither prime nor composite"
    else:
        for i in range(2, int(n**0.5)+1):
            if n % i == 0:
                primality = "Composite"
                break
        else:
            primality = "Prime"

    # Small or large
    if n <= 10:
        size = "Small"
    else:
        size = "Large"

    # Combine
    return f"{n} → {parity}, {primality}, {size}"


print(describe_number(9))
print(describe_number(2))
print(describe_number(16))
print(describe_number(11))

import seaborn as sns
titanic = sns.load_dataset("titanic").copy()

print(titanic.head(5))

#filter with loc
female_first = titanic.loc[(titanic["sex"] == "female") & (titanic["class"] == "First")]
older_than40 = titanic.loc[titanic["age"] > 40]
more_than_2sibsp = titanic.loc[titanic["sibsp"] > 2]

print(female_first.head())
print(older_than40.head())
print(more_than_2sibsp.head())

titanic["family_size"] = titanic["sibsp"] + titanic["parch"] + 1

#mean age and survival rate grouped by sex and class
mean_age_survival = titanic.groupby(["sex","class"])[["age","survived"]].mean()

#top 10 passengers by fare
top10_byfare = titanic.sort_values("fare", ascending=False).head(10)

print("Mean age survival: \n",mean_age_survival.head(10))
print("Top 10: \n",top10_byfare)

import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(data=titanic, x="class", y="survived", hue="sex")
plt.title("Survival rate by class and sex")
plt.tight_layout()
plt.show()
```
