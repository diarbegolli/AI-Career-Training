```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset: players with goals and assists
players = pd.DataFrame({
    "player": ["Haaland","Messi","Ronaldo","Mbappe","Kane","Benzema","Lewandowski","Neymar"],
    "team": ["Man City","PSG","Man Utd","PSG","Spurs","Real Madrid","Barcelona","PSG"],
    "goals": [36,30,25,28,22,20,23,19],
    "assists": [5,12,6,9,8,7,4,11],
    "season": ["2022/23"]*8
})
print(players)

sns.boxplot(data=players, x="team", y="goals")
plt.xticks(rotation=45)
plt.title("Goals distributed by team")
plt.show()

sns.violinplot(data=players,x="team", y="assists")
plt.xticks(rotation=45)
plt.title("Assist distribution by team")
plt.show()

sns.regplot(data=players, x="goals", y="assists")
plt.title("Goals vs Assists(with regression)")
plt.show()

sns.scatterplot(data=players, x="goals", y="assists", hue="team", style="team", s=150)
plt.title("goals vs assists by team")
plt.show()

sns.pairplot(players[["goals","assists"]])
plt.suptitle("Pairplot of Goals & Assists", y=1.02)
plt.show()
```
